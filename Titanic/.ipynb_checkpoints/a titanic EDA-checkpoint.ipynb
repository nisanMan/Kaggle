```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#Accepts a DataFrame and returns a graph displaying NaN values.
def nan_values(df):
    ax = plt.axes()
    sns.heatmap(df.isna().transpose(), cbar=False, ax=ax)
    plt.title('Missing Values', fontsize=12)
    plt.xlabel('PassengerId', fontsize = 10) 
    plt.ylabel('Columns', fontsize = 10)
    plt.show()
```

# EDA Name --> Title

```python
#from name --> Title and drop the name 
titanic["Title"] = titanic["Name"].str.extract("([A-Za-z]+)\.",expand=True)
titanic.drop(columns=['Name'], inplace=True)
```

# Explore the 'Title' column


```python
df = titanic.copy()[['Title','Survived']].groupby(by=[titanic['Title']]).agg({'Title':'count', 'Survived':'sum'}).rename(columns={'Title': 'Count Title'})
df
```

```python
# Mapping titles to the specified groups
'''
1.Crew Members of a Ship: Capt, Col, Major
2.Nobility: Countess, Don, Jonkheer, Lady, Sir
3.Thorn Men: Master, Mr
4.Women: Miss, Mlle, Mme, Mrs, Ms
5.Professions: Dr, Rev
'''
title_mapping = {
    'Capt': 'Crew Members of a Ship',
    'Col': 'Crew Members of a Ship',
    'Major': 'Crew Members of a Ship',
    'Countess': 'Nobility',
    'Don': 'Nobility',
    'Jonkheer': 'Nobility',
    'Lady': 'Nobility',
    'Sir': 'Nobility',
    'Master': 'Thorn Men',
    'Mr': 'Thorn Men',
    'Miss': 'Women',
    'Mlle': 'Women',
    'Mme': 'Women',
    'Mrs': 'Women',
    'Ms': 'Women',
    'Dr': 'Professions',
    'Rev': 'Professions'
}

# Rename the index based on the mapping
df.rename(index=title_mapping, inplace=True)
```

```python
df = df.groupby(by=df.index).sum()
df
```

```python
titanic['Title'] = titanic['Title'].replace(title_mapping)
```

# EDA Ticket --> Ticket_numbers(NaN ->0) , Ticket_letters (NaN ->ZZZ)

```python
# Extract letters and numbers
titanic['Ticket_letters'] = titanic['Ticket'].str.extract(r'([A-Za-z]+)')
titanic['Ticket_numbers'] = titanic['Ticket'].str.extract(r'(\d+)')

# Drop the original Ticket column
titanic.drop(columns=['Ticket'], inplace=True)
```

```python
df = titanic.copy()[['PassengerId','Survived','Ticket_numbers','Ticket_letters']]
df['Ticket_letters'] = df['Ticket_letters'].fillna('ZZZ')
df
```

### EDA Ticket_letters

```python
total = df.groupby('Ticket_letters').agg({'PassengerId':'count', 'Survived':'sum'})
total['%s']=100*total['Survived']/total['PassengerId']
total
```

```python

# Data
index = total.index
percentage = total['%s']

# Create bar plot
plt.figure(figsize=(12, 6))
plt.bar(index, percentage, color='blue')
plt.xlabel('Ticket_letters')
plt.ylabel('%s')
plt.title('Percentage by Ticket_letters')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Filter the data
filtered_data = total[total['PassengerId'] > 9]

# Extract index and percentage
index = filtered_data.index
percentage = filtered_data['%s']

# Create bar plot
plt.figure(figsize=(12, 6))
plt.bar(index, percentage, color='blue')
plt.xlabel('Ticket_letters')
plt.ylabel('%s')
plt.title('Percentage by Ticket_letters')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### EDA Ticket_numbers

```python
total = titanic.copy()[['PassengerId','Survived','Ticket_numbers','Ticket_letters']]

total['Ticket_numbers'] = total['Ticket_numbers'].fillna(0) #in total fill 0 bat in titanic fill 55000
total['Ticket_numbers'] = total['Ticket_numbers'].astype(int)
total['Ticket_numbers']
```

```python
bin_edges = np.linspace(0, 370376, 50)  # Adjust the number of bins as needed

# Plot histogram
plt.figure(figsize=(12, 6))
plt.hist(total['Ticket_numbers'] ,bins=bin_edges, color='r', alpha=0.7)
plt.hist(total[total['Survived']==1]['Ticket_numbers'] ,bins=bin_edges, color='blue', alpha=0.7)
plt.xlabel('Ticket_numbers')
plt.ylabel('Frequency')
plt.title('Histogram of Ticket_numbers blue is Survived')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.ticklabel_format(style='plain', axis='x')  # Display x-axis labels in plain format
plt.show()
```

```python
bins = [0, 50000, 100000, 200000, 300000, 400000]

# Create labels for the bins
labels = ['0-50000', '50000-100000', '100000-200000', '200000-300000', '300000-400000']

# Create a new column 'Ticket_bins' based on the bins
total['Ticket_bins'] = pd.cut(total['Ticket_numbers'], bins=bins, labels=labels, right=False)

# Group by 'Ticket_bins' and 'Survived' and count
grouped = total.groupby(['Ticket_bins', 'Survived']).size().unstack(fill_value=0)

# Plot stacked histogram
grouped.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Ticket_numbers Range')
plt.ylabel('Frequency')
plt.title('Distribution of Ticket_numbers by Survived')
plt.grid(True)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Cast data like 'final mail' to 0 and 1

column_names = [
 'Embarked',
 'Title',
 'Cabin_letters',
 'Sex'
]
def set_catagoric(cul_name):
    le_cul_name = LabelEncoder()
    le_cul_name.fit(titanic[cul_name])

    encoded_cul_name = le_cul_name.transform(titanic[cul_name])
    titanic[cul_name] = encoded_cul_name
for column_name in  column_names:
    
    set_catagoric(column_name)



titanic.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Split the data into train and test sets
X = df.drop(columns=['Survived'])
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

y_train =  y_train.astype(int)
y_test = y_test.astype(int)


sc =StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
beta = LogisticRegression(random_state=65,penalty=None, max_iter=10000)
beta.fit(X_train,y_train)
beta.intercept_,beta.coef_
```

```python
X.head()
```

```python
y_pred = beta.predict(X_train)

accuracy_score(y_train, y_pred) 
```

```python
y_pred = beta.predict(X_test)

accuracy_score(y_test, y_pred) 
```

```python
from sklearn.metrics import confusion_matrix



# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:\n", cm)
```
